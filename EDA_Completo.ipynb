```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Amazon = pd.read_csv("./Datasets/amazon_prime_titles.csv", sep=",")
Disney = pd.read_csv("./Datasets/disney_plus_titles.csv", sep=",")
Hulu = pd.read_csv("./Datasets/hulu_titles.csv", sep=",")
Netflix = pd.read_csv("./Datasets/netflix_titles.csv", sep=",")
Plataformas_ratings= pd.read_csv("./plataformas_ratings.csv", sep=",")
```

```python
Amazon['source'] = 'Amazon'
Disney['source'] = 'Netflix'
Hulu['source'] = 'Hulu'
Disney['source'] = 'Disney'
```

```python
Plataformas = pd.concat([Amazon, Netflix, Hulu, Netflix])
```

Graficos para analizar

```python
sns.countplot(x='type', data=Netflix)
plt.title('Distribución de tipos de contenido en Netflix')
plt.show()
```

```python
sns.countplot(x='rating', data=Amazon)
plt.title('Distribución de clasificaciones de Amazon')
plt.show()
```

```python
sns.histplot(x='release_year', data=Hulu)
plt.title('Distribución de años de lanzamiento en Hulu')
plt.show()
```

```python
Disney_rating = Disney['rating'].value_counts()
plt.pie(Disney_rating, labels=Disney_rating.index, autopct='%1.1f%%')
plt.title('Distribución de clasificaciones de Disney')
plt.show()
```

```python
sns.countplot(x='source', hue='type', data=Plataformas)
plt.title('Distribución de tipos de contenido por fuente')
plt.show()
```

```python
sns.boxplot(x='type', y='duration_int', data=Plataformas_ratings)
plt.title('Distribución de la duración en todas las plataformas')
plt.show()
```

```python
#Buscamos valores atípicos
outliers = Plataformas_ratings['duration_int'][Plataformas_ratings['duration_int'] > Plataformas_ratings['duration_int'].quantile(0.75) + 1.5*(Plataformas_ratings['duration_int'].quantile(0.75) - Plataformas_ratings['duration_int'].quantile(0.25))].values
print("Valores atípicos:", outliers)
```

```python
#histograma de scores totales
sns.histplot(data=Plataformas_ratings, x="score_mean")
plt.title('Distribución de una variable en el score')
plt.show()
```
