```python
Plataformas = pd.concat([Amazon, Disney, Hulu, Netflix], ignore_index=True)
```

```python
Plataformas.head(3)
```

```python
Plataformas = Plataformas.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
Plataformas.head(3)
```

```python
Ratings_df = pd.concat([pd.read_csv(file) for file in list_csv])
Ratings_df = Ratings_df.rename(columns={"movieId": "id"})
Plataformas = pd.merge(Plataformas, Ratings_df, on='id', how='inner')
```

```python
promedios = Ratings_df.groupby("id")["rating"].mean()
promedios_round = promedios.round(2)
score_mean = pd.DataFrame({'id': promedios_round.index, 'score_mean': promedios_round})
score_mean.reset_index(drop=True, inplace=True)
```

```python
Plataformas = pd.merge(Plataformas,score_mean,on='id', how='inner')
```

```python
Plataformas = Plataformas.drop("rating_y", axis=1)
```

```python
Plataformas = Plataformas.drop("userId", axis=1)
```

```python
Plataformas = Plataformas.drop("timestamp", axis=1)
```

```python
Plataformas = Plataformas.drop_duplicates()
```
